# Create a score for one address

```python
import os
import sys


PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)
```

```python
import pandas as pd

from src.features import extract_location_features
from src.geo import geocode_address
from src.scoring import score_location
```

## Load address and POIs

```python
poi_data = pd.read_csv('../data/normalized_pois.csv')
poi_data.head()
```

```python
# Get address from environment variable
ADDRESS = os.getenv('ADDRESS')
if not ADDRESS:
    raise ValueError(
        'Please set the ADDRESS environment variable to the location you want to score.'
    )
```

```python
# Geocode the address to get lat and lon values
address_coords = geocode_address(ADDRESS)
if address_coords:
    address_lat, address_lon = address_coords
```

## Extract features for a location

```python
# Extract features for the given location
features = extract_location_features(address_lat, address_lon, poi_data)
features
```

## Calculate score

```python
# Score the location based on the extracted features
result = score_location(features)
result
```
