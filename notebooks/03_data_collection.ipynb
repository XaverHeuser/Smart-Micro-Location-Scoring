# Data Collection from Overpass API

```python
import os
import sys


PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)
```

```python
from src.config import DEFAULT_OVERPASS_URL, POINTS_OF_INTEREST, QUALITY_METRICS
from src.geo import geocode_address
from src.overpass_api import fetch_poi_data
from src.poi_processing import normalize_pois
```

```python
ADDRESS = os.getenv('ADDRESS')

print(f'Using Overpass API URL: {DEFAULT_OVERPASS_URL}')
print(f'Points of Interest: {POINTS_OF_INTEREST}')
print(f'Quality Metrics: {QUALITY_METRICS}')
```

## Get coords of address

```python
address_coords = geocode_address(ADDRESS)
if address_coords:
    address_lat, address_lon = address_coords
```

## Fetch POIs from Overpass

```python
raw_poi_data = fetch_poi_data(
    lat=address_lat,
    lon=address_lon,
    categories=POINTS_OF_INTEREST,
    radius=5000,
    overpass_url=DEFAULT_OVERPASS_URL,
)
```

```python
print(f'Fetched {len(raw_poi_data)} POIs from Overpass API.')
```

## Normalize POIs data

```python
poi_df = normalize_pois(raw_poi_data)
print(f'Len of normalized POI DataFrame: {len(poi_df)}')
```

## Save POIs

```python
poi_df.to_csv('../data/normalized_pois.csv', index=False)
```
